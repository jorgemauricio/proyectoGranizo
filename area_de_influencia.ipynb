```python
# librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import math
%matplotlib inline
```

```python
dataWrf = pd.read_csv("data/Resultado_wrf_1.csv")
```

```python
dataWrf.info()
```

```python
def generar_area_de_influencia(v):
    for i in range(0,8):
        value = v * math.sin(random.randint(0,99)/100)
        v += value
    return v
```

```python
dataWrf["areaInfluencia"] = dataWrf.apply(lambda x: generar_area_de_influencia(x["RainWRF"]), axis=1)
```

```python
dataWrf.info()
```

```python
dataWrf[["RainWRF", "areaInfluencia"]].sum()
```

```python
dataWrf.to_csv("data/Resultado_wrf_area_1.csv")
```
