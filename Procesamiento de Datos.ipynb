```python
dataTemp = data[(data["Year"] == 2014) & (data["Month"] == 8) & (data["Day"] == 20) & (data["Nombre"] == "Vista Hermosa Victor")]
```

```python
dataTemp.index = dataTemp["Fecha"]
dataTemp.index = pd.to_datetime(dataTemp.index)
```

```python
fig = plt.figure(figsize=(24,5))
ax = fig.add_subplot(111)

x = np.array(dataTemp.index.hour)
y = np.array(dataTemp["RainIMR"])
e = np.array(dataTemp["validacionEvento"])

line, = ax.plot(x, y, lw=2)

ax.set_xlabel("Hora")
ax.set_ylabel("Precipitación mm")
ax.set_title("Vista Hermosa Victor")

ax.annotate('Evento', xy=(x[e.argmax()], y.max()), xytext=(x[e.argmax()]+1, y.max()), arrowprops=dict(facecolor='red', shrink=0.02))

plt.show()
```
